```python
import pandas as pd 
import json

docs = pd.read_json("parsed_docs.json")

docs.head()
```

Videos are absent and parser always fetch logo

```python
docs = docs[["url","title","text","images"]]
docs.head()
```

```python
unwanted_image = "https://files.readme.io/fbf05e1-4370cbd-bitmovin-logo-horizontal-docs-black.svg"

docs["images"] = docs["images"].apply(
    lambda img_list: [img for img in img_list if img != unwanted_image]
)
```

```python

with open("parsed_docs_cleaned.json", "w", encoding="utf-8") as f:
    json.dump(docs.to_dict(orient="records"), f, indent=2, ensure_ascii=False)
```
